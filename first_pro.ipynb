# feature scaling

# Soil_pH 

```python
#Soil_pH 
ss=StandardScaler()
ss.fit(dataset[["Soil_pH"]])
```

```python
dataset['Soil_pH_ss']=pd.DataFrame(ss.transform(dataset[["Soil_pH"]]),columns=['x'])
```

# comparision

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["Soil_pH"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["Soil_pH_ss"])
```

# Rainfall_mm

```python
#Rainfall_mm
ss=StandardScaler()
ss.fit(dataset[["Rainfall_mm"]])
```

```python
dataset['Rainfall_mm_ss']=pd.DataFrame(ss.transform(dataset[["Rainfall_mm"]]),columns=['x'])
```

# comparision

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["Rainfall_mm"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["Rainfall_mm_ss"])
```
